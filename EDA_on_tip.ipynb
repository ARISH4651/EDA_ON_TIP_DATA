#EDA ANALYSIS ON TIP DATA

**Firstly import neccessay libs for EDA**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Here we take inbuilt datasets from seaborn**

```python
tip_data=sns.load_dataset('tips')
```

**Our dataset contains:**

INDEPENDENT FEATURES:

Total bill

Sex

Smoker

Day

Time

Size

DEPENDENT FEATURES:

Tip

```python
tip_data.head()
```

**Shape is used to know about size of datasets**

```python
tip_data.shape
```

**isnull() is used to check missing values in datasets**

```python
tip_data.isnull().sum()
```

**Visualizing whether datasets contains missing values using heatmap**

```python
sns.heatmap(tip_data.isnull() , cbar=False , cmap='viridis')
```

**Correlation:-Used to get to know relationship between two features**

Nearly to 1:- Positive correlation

Neraly to -1:- Negative correlation

Zero :- no correlation

```python
corr = tip_data[['total_bill', 'tip', 'size']].corr()
```

**Visualizating correlation through heatmap**

```python
sns.set()
plt.figure(figsize=(6,6))
sns.heatmap(corr , square=True , annot=True , fmt='.1f',cbar=True , cmap='plasma')
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='sex' , data=tip_data , color='skyblue')
plt.xlabel('SEX')
plt.ylabel("Count per sex")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='smoker' , data=tip_data , palette='Purples')
plt.xlabel('SOMKER')
plt.ylabel("COUNT PER SMOKER")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.distplot(tip_data['tip'], kde=False)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='day', hue='smoker',data=tip_data, color='blue')
plt.xlabel('DAY')
plt.ylabel("smoker")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='sex', hue='smoker', data=tip_data, palette='viridis')
plt.xlabel('SEX')
plt.ylabel("SMOKER")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='sex', hue='time', data=tip_data, palette='plasma')
plt.xlabel('SEX')
plt.ylabel("TIME")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='sex', hue='size', data=tip_data)
plt.xlabel('SEX')
plt.ylabel("SIZE")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.jointplot(x='total_bill' , y='tip' , data=tip_data)
plt.xlabel('TOTAL BILL')
plt.ylabel('TIP')
plt.show()
```
